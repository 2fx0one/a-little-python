# 简单线性回归

```python
x = np.array([1., 2., 3., 4., 5.])
y = np.array([1., 3., 2., 3., 5.])
```

```python
plt.scatter(x,y)
plt.axis([0, 6, 0, 6])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

#Numerator and denominator
n = 0.
d = 0.
for x_i, y_i in zip(x, y):
    n += (x_i - x_mean) * (y_i - y_mean)
    d += (x_i - x_mean)**2
    
a = n/d

b = y_mean - a * x_mean

print(a, b)
```

```python
y_hat = a * x + b

plt.scatter(x, y)
plt.plot(x, y_hat, color='r')
plt.axis([0, 7, 0, 7])
plt.show()
```

```python
x_input = 6
y_output = a * x_input + b
print(y_output)
```
