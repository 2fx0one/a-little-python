# 向量化运算

### w * v 两个向量点乘。就是向量各维相乘，再相加求和。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1., 2., 3., 4., 5.])
y = np.array([1., 3., 2., 3., 5.])
```

```python
plt.scatter(x,y)
plt.axis([0, 6, 0, 6])
plt.show()
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)

n = (x - x_mean).dot(y - y_mean)
d = (x - x_mean).dot(x - x_mean)
```

```python
a = n / d
b = y_mean - a * x_mean
print(a, b)
```

```python
y_hat = a * x + b

plt.scatter(x, y)
plt.plot(x, y_hat, color='r')
plt.axis([0, 7, 0, 7])
plt.show()
```

```python
x_input = 6
y_output = a * x_input + b
print(y_output)
```

```python
m =  1_00_0000
x = np.random.random(size=m)
y = x * 2.0 + 5.0 + np.random.normal(size=m)

def reg_vector(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    n = (x - x_mean).dot(y - y_mean)
    d = (x - x_mean).dot(x - x_mean)
    a = n / d
    b = y_mean - a * x_mean
    print(a, b)
    
def reg_simple(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)

    #Numerator and denominator
    n = 0.
    d = 0.
    for x_i, y_i in zip(x, y):
        n += (x_i - x_mean) * (y_i - y_mean)
        d += (x_i - x_mean)**2

    a = n/d

    b = y_mean - a * x_mean

    print(a, b)
```

```python
%time reg_simple(x, y)
```

```python
%time reg_vector(x, y)
```
